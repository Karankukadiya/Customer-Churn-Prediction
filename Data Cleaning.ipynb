```python
import pandas as pd
import numpy as np
```

# Load the dataset

```python
df = pd.read_csv('Original Dataset.csv')
```

```python
print(f"Shape: {df.shape}")         
print(df.info())
print(df.head())
```

# Remove duplicate records

```python
print(f"Duplicates found: {df.duplicated().sum()}")  
```

```python
df = df.drop_duplicates(keep='first').reset_index(drop=True)
```

```python
print(f"After removing duplicates: {df.shape}") 
```

# Drop rows with null values

```python
print(f"Nulls per column:\n{df.isnull().sum()}")
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
print(f"After dropping nulls: {df.shape}")
```

# Standarize categorical columns

```python
# Fix Gender
gender_map = {
    'male': 'Male', 'M': 'Male',
    'female': 'Female', 'F': 'Female',
    'other': 'Other'
}
df['Gender'] = df['Gender'].replace(gender_map)

# Fix PlanType
plan_map = {
    'prepaid': 'Prepaid',
    'POSTPAID': 'Postpaid', 'postpaid': 'Postpaid'
}
df['PlanType'] = df['PlanType'].replace(plan_map)

# Fix Churn & PortRequest
yn_map = {
    'yes': 'Yes', 'YES': 'Yes', '1': 'Yes',
    'no': 'No',  'NO': 'No',  '0': 'No'
}
df['Churn'] = df['Churn'].replace(yn_map)
df['PortRequest'] = df['PortRequest'].replace(yn_map)

print(df['Gender'].value_counts())
print(df['Churn'].value_counts())
```

# Remove out of range invalid inputs

```python
df = df[
    (df['Age'] >= 18) & (df['Age'] <= 99) &
    (df['SatisfactionScore'] >= 1) & (df['SatisfactionScore'] <= 5) &
    (df['RechargeAmount'] > 0) &
    (df['SIMTenureMonths'] > 0) &
    (df['Gender'].isin(['Male', 'Female', 'Other'])) &
    (df['PlanType'].isin(['Prepaid', 'Postpaid'])) &
    (df['Churn'].isin(['Yes', 'No'])) &
    (df['PortRequest'].isin(['Yes', 'No']))
].reset_index(drop=True)

print(f"After validation: {df.shape}")
```

# Fix data types

```python
int_cols = [
    'Age', 'SIMTenureMonths', 'RechargeAmount',
    'RechargeFrequencyDays', 'CallMinutes', 'SMSUsage',
    'NetworkComplaints', 'DroppedCalls', 'LateRechargeDays',
    'CustomerSupportCalls', 'SatisfactionScore'
]

for col in int_cols:
    df[col] = df[col].astype('int64')

print(df.dtypes)
```

# save clean data

```python
df.to_csv('Dataset.csv', index=False)
```
